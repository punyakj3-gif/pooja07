```python
import pandas as pd
import numpy as np
data={
    'First':['Arujan','Rohan','Riya','Yash','siddini'],
    'Lastname':['Singli','Agarwal','Shah','Batia','Khan',],
    'Type':['Full_time employee','Intern','Full_time employee','part time employee','Full_time employee'],
    'Department':['Administration','technical','Administration','Technical','Technical'],
    'yoe':[2,3,5,7,6],
    'salary':[20000,2000,1000,2000,4000]
}
df=pd.DataFrame(data)
print(df)
table=pd.pivot_table(df,index=['Type','Department'],values='salary',aggfunc=['mean'])
print(table)
table1=pd.pivot_table(df,index=['Type','Department'],values='salary',aggfunc=['sum','mean','count'])
print(table1)
table2=pd.pivot_table(df,index=['Type'],values='salary',aggfunc=['count'])
print(table2)
table3=pd.pivot_table(df,index=['Type'],values='salary',aggfunc=['std'])
print(table3)
```
